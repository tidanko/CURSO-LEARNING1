```python
from matplotlib import pyplot as plt
import numpy as np
pvc = plt.imread('./pikachu_vs_charmander.jpeg')
```

## Lightness method

```python
plt.imshow((pvc.max(axis = 2) + pvc.min(axis = 2)) / 2, cmap='gray', vmin=0, vmax=255)
```

## Average method

```python
plt.imshow((pvc[:,:, 0] + pvc[:,:, 1] + pvc[:,:, 2]) / 3, cmap='gray', vmin=0, vmax=255)
```

## Luminosity method

```python
plt.imshow(0.3 * pvc[:,:, 0] + 0.59 * pvc[:,:, 1] + 0.11 * pvc[:,:, 2], cmap='gray', vmin=0, vmax=255)
```
