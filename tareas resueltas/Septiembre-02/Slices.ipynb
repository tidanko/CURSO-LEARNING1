## Lectura del logo

```python
from matplotlib import pyplot as plt
logo = plt.imread('./AFALogo.bmp')
```

### Ejercicio 1

```python
print(logo.shape)
plt.imshow(logo[80:155, 25:103], cmap='gray', vmin=0, vmax=255.)
```

### Ejercicio 2

```python
plt.imshow(logo[0:33, :].T, cmap='gray', vmin=0, vmax=255.)
```

### Ejercicio 3

```python
logo_copia = logo.copy()
logo_copia[logo_copia > 0] = 255
plt.imshow(logo_copia, cmap='gray', vmin=0, vmax=255.)
```

### Ejercicio 4

```python
logo_copia_2 = logo.copy()
logo_copia_2[logo_copia_2 < 150] = 255
plt.imshow(logo_copia_2, cmap='gray', vmin=0, vmax=255.)
```

### Ejercicio 5

```python
plt.imshow(logo[:, ::-1], cmap='gray', vmin=0, vmax=255.)
```

### Ejercicio 6

```python
logo_copia_2[:, ::5] = 0
logo_copia_2[::5, :] = 0
plt.imshow(logo_copia_2, cmap='gray', vmin=0, vmax=255.)
```
